```python
def number_of_injuries(row):
    #this returns the number of games skipped for injuries in a row
    total_list=row['0'].strip('[]').split(', ')
    row['injured games']=sum( [y.strip("''").strip('\\xa0') for y in total_list].count(x) for x in true_injuries) 
    return row
```
