```python
playerLinksITA = []

#Run the scraper through each of our 20 team links
for i in range(len(teamLinksITA)):

    #Download and process the team page
    page = teamLinksITA[i]
    tree = requests.get(page, headers = headers)
    soup = BeautifulSoup(tree.content, 'html.parser')

    #Extract all links
    links = soup.select("a.spielprofil_tooltip")
    
    #For each link, extract the location that it is pointing to
    for j in range(len(links)):
        playerLinksITA.append(links[j].get("href"))
        
#The page list the players more than once - let's use list(set(XXX)) to remove the duplicates
playerLinksITA = list(set(playerLinksITA))

#Add the location to the end of the transfermarkt domain to make it ready to scrape
for j in range(len(playerLinksITA)):
    playerLinksITA[j] = "https://www.transfermarkt.co.uk"+playerLinksITA[j]
```

```python
PlayerSerieAlinks=[]
for x in playerLinksITA:
    PlayerSerieAlinks.append(x.split('/profil/')[0]+'/leistungsdatendetails/'+x.split('/profil/')[1]+'/plus/0?saison=2016&verein=&liga=&wettbewerb=&pos=&trainer_id=')
```

```python
names=[]
injuries=[]

for i in range(len(PlayerSerieAlinks)):

    #Take site and structure html
    page = PlayerSerieAlinks[i]
    tree = requests.get(page, headers=headers)
    soup = BeautifulSoup(tree.content, 'html.parser')
    
    names.append(soup.find("a", {"class": "spielprofil_tooltip"})["title"])

    rows=soup.find_all("tr", {"class": "bg_rot_20"})
    issues=[]
    for row in rows:
        issues.append(row.find('td', {"colspan": 8}).text)
    injuries.append(issues)
```
