```python
def number_of_injuries(row):
    #this returns the number of games skipped for injuries in a row
    total_list=row['injuries'].strip('[]').split(', ')
    row['injured games']=sum( [y.strip("''").strip('\\xa0') for y in total_list].count(x) for x in yes_injuries_list) 
    return row
```

```python
def lastandmean(row):
    week=int(row['injury week'])
    row['last_variation']=None
    row['mean_variation']=None
    if week ==1:
        return row
    if week ==2:
        row['mean_variation']=row['mean']-row[1]
        return row
    row['mean_variation']=row['mean']-row[week-1]
    row['last_variation']=row[week-1]-row[week-2]
    return row
```
