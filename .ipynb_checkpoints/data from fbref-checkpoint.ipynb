```python
import numpy as np
import pandas as pd
import pickle
```

```python
with open("linkslist.txt", "rb") as fp:   # Unpickling
    linkslist = pickle.load(fp)
```

```python
cols_to_drop=['Date','Day','Comp','Venue','Result','Squad','Opponent',
              'Start','Pos','Gls','Ast','PK','PKatt','CrdY','CrdR','Match Report']
row_list=[]

for link in linkslist:

    '''returns a dataframe with the data of the player for the Premier League 2017/18'''
    player_data=pd.read_html(link,header=1)[0]
    
    if 'GK' in player_data['Pos'].values:
        continue #we do not add rows because the players is a goalie
    
    #player_data=player_data[player_data['Comp']=='Premier League'] #we only keep PL games
    player_data.drop(cols_to_drop, axis=1, inplace=True, errors='ignore')
    #pl_data=pl_data[pl_data['Min'].map(lambda x: len(x)<=2)] #we only keep rows where the player played the game

    player_data.insert(0,'name',link.split('/')[-1].split('Match')[0])    
    #pl_data.set_index('name',inplace=True)
    print(link)
    
    row_list+=player_data.to_numpy().tolist()
```

```python
with open("fbrefdata.txt", "wb") as fp:   #Pickling
    pickle.dump(row_list, fp)
```

```python
all_players_data=pd.DataFrame(good_list, columns=my_columns)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
#Process League Table for La Liga season 2017/18 to get the teams names
page = 'https://fbref.com/en/comps/12/1652/2017-2018-La-Liga-Stats'
tree = requests.get(page, headers = headers)
soup = BeautifulSoup(tree.content, 'html.parser')

#Create an empty list to assign these values to
teamLinks = []

#Extract all links with the correct CSS selector
rows=soup.find_all("td", {"class": "left ", "data-stat":"squad"})

#We need the location that the link is pointing to, so for each link, take the link location. 
#Additionally, we only need the links in locations 1,4,7,etc. of our list, so loop through those only
for row in rows:
    teamLinks.append('https://fbref.com/'+row.find("a").get("href"))
```

```python
teamLinks=teamLinks[:20]
```

```python
#Create an empty list for our player links to go into
playerLinks = []

#Run the scraper through each of our 20 team links
for i in range(len(teamLinks)):

    #Download and process the team page
    page = teamLinks[i]
    tree = requests.get(page, headers = headers)
    soup = BeautifulSoup(tree.content, 'html.parser')

    table = soup.find("tbody")
    playerurls=table.find_all("td", {"class":"left group_start","data-stat":"matches"}) 
    
    playerLinks+=[url.find("a").get("href") for url in playerurls]
```

```python
cols_to_drop=['Date','Day','Comp','Venue','Result','Squad','Opponent',
              'Start','Pos','Gls','Ast','PK','PKatt','CrdY','CrdR','Match Report']
row_list=[]

for link in playerLinks:

    '''returns a dataframe with the data of the player for the Premier League 2017/18'''
    player_data=pd.read_html(link,header=1)[0]
    
    if 'GK' in player_data['Pos'].values:
        continue #we do not add rows because the players is a goalie
    
    #player_data=player_data[player_data['Comp']=='Premier League'] #we only keep PL games
    player_data.drop(cols_to_drop, axis=1, inplace=True, errors='ignore')
    #pl_data=pl_data[pl_data['Min'].map(lambda x: len(x)<=2)] #we only keep rows where the player played the game

    player_data.insert(0,'name',link.split('/')[-1].split('Match')[0])    
    #pl_data.set_index('name',inplace=True)
    print(link)
    
    row_list+=player_data.to_numpy().tolist()
```
