```python
# this function separates the full name in the injuries dataframe into two separate columns for first and last name
def split_names(row):
    if row['0'].count(' ')==0:
        row['first']=row['0']
        row['last']=None
    else:
        row['first'],row['last']=row['0'].split(' ',1)
    return row
```

```python
def std_pass_percentage(row):
    data = row[weeks]
    row['std'] = np.nanstd(data)
    return row
```

```python
passaccuracysample=passaccuracysample.apply(std_pass_percentage, axis=1)
```

```python
passaccuracysample.plot.scatter(x='avg',y='std')
```

```python
df2=pd.merge(passaccuracysample[['playerId','avg','std']],injuries,how='inner',left_on=['last','first'], right_on=['last','first'])
df2.head()
```

```python
len(df2)
```

```python
df2=df2.drop(columns=['0'])
```

```python
df2.head()
```

```python
def injurytype(player):
    if '\xa0Knee Injury' in player.values:
        return 'Knee'
    elif '\xa0Hamstring Injury' in player.values:
        return 'Hamstring'
    elif '\xa0Ankle Injury' in player.values:
        return 'Ankle'
    elif '\xa0Back Injury' in player.values:
        return 'Back'
    else:
        return 'Other'
```

```python
def injury(row):
    row['injury_type']=injurytype(row)
    return row
```

```python
df2=df2.apply(injury, axis=1)
```

```python
df2.head()
```

```python
superplot = plt.figure(figsize=(10, 10))
colors = ['#1abc9c', '#ff0084', '#3498db', '#f1c40f', '#34495e']

for color, (lab, x) in zip(colors, df2.groupby('injury_type')):
    plt.plot(x['avg'], x['std'], 'o', label=lab, color=color)
plt.xlabel('Average Percentage')
plt.ylabel('Standard Deviation')
plt.legend()
```
